1. Introducción:

En esta sección se proporciona una breve introducción al análisis realizado a tres empresas que ofrecen cursos en línea en diferentes modalidades, en respuesta a la creciente demanda de los cursos MOOCs (Cursos Masivos Abiertos en Línea, por sus siglas en inglés) en diversos entornos académicos. Se destaca la importancia de estos cursos para la difusión masiva de información y el acceso global a la misma.

2. Objetivos:

El objetivo principal de este análisis es obtener una comprensión detallada del comportamiento de la oferta y la demanda de los cursos MOOCs, mediante el estudio de diferentes conjuntos de datos. Esto incluye la segmentación del mercado, los canales de distribución utilizados, las tendencias identificadas y la tasa de conversión. A partir de esta información, se busca identificar aspectos clave del negocio y plantear una propuesta coherente que se alinee con las necesidades del mercado actual.

El análisis de los datasets permitirá examinar en profundidad la información relacionada con los cursos MOOCs, como la cantidad de usuarios registrados, las áreas temáticas más populares, los patrones de consumo de contenido, la duración promedio de los cursos, los niveles de participación entre otros indicadores relevantes. Asimismo, se evaluarán los canales de distribución utilizados para promover y comercializar los cursos MOOCs como asociaciones con instituciones educativas y otras estrategias de difusión. Esto ayudará a comprender cómo se llega a los usuarios y cómo se genera la demanda de los cursos.

El análisis también se centrará en las tendencias observadas en el mercado de los cursos MOOCs, como el crecimiento de ciertas áreas temáticas, cambios en los perfiles de los usuarios, avances tecnológicos que impactan en la experiencia de aprendizaje, entre otros factores relevantes. Estas tendencias proporcionarán información valiosa para adaptar la propuesta de cursos a las necesidades emergentes.

Con base en el análisis de estos aspectos clave del negocio, se buscará plantear una propuesta coherente que responda a las necesidades y demandas del mercado actual de los cursos MOOCs. Esto incluirá recomendaciones para mejorar la oferta de cursos, estrategias de marketing más efectivas, mejoras en la experiencia de usuario y posibles oportunidades de crecimiento.


3. Metodología:

Dentro de la metodología propuesta, se comienza realizando la carga de cada uno de los archivos CSV en Python, con el objetivo de garantizar que la información proporcionada cumpla con los requisitos mínimos para iniciar el proceso de análisis exploratorio de datos. En esta primera etapa, se realiza un proceso de extracción, transformación y carga (ETL), asegurándose de dejar el dataset en un formato adecuado y listo para su análisis.

Una vez que se ha completado la etapa de carga y limpieza de los datos, se procede al análisis exploratorio de datos. Esta fase consiste en examinar detalladamente el conjunto de datos y realizar diversas operaciones para obtener información relevante.  Se aplciaran distintas técnicas de estadística descriptiva, visualización de datos y análisis de tendencias para comprender la distribución de los datos, identificar patrones, correlaciones y outliers, y obtener una visión general de la información contenida en el dataset.

El análisis exploratorio de datos se realizo por separado a cada uno de los dataset y los archivos tiene los siguientes nombres:

Esto permitirá visualizar de manera clara y concisa los aspectos más relevantes y proporcionará una visión general de los datos analizados.
